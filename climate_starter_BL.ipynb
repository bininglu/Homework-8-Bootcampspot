# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
# session = Session(engine)
conn = engine.connect()
session = Session (bind = conn)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results


# Calculate the date 1 year ago from the last data point in the database
all_prcp = session.query(Measurement).order_by(desc(Measurement.date)).limit(1)
for prcp in all_prcp:
    last_date=prcp.date

one_year_ago=dt.datetime.strptime(last_date, '%Y-%m-%d').date()- dt.timedelta(days=365)
print(last_date)
print(one_year_ago)

# Perform a query to retrieve the data and precipitation scores
select_db = session.query(Measurement.date, Measurement.prcp).\
            filter(Measurement.date >= one_year_ago).\
            order_by(Measurement.date).all()
# print(select_db)
# Save the query results as a Pandas DataFrame and set the index to the date column
select_df = pd.DataFrame(select_db, columns=['Date', 'Precipitation']).set_index('Date')

# Sort the dataframe by date
select_df=select_df.sort_values('Date')
select_df=select_df.reset_index()
# print(select_db)
x_axis = np.arange(len(select_df))
plt.bar(select_df['Date'],select_df['Precipitation'], align='center', label="Precipitation")
plt.xticks([])
plt.xlabel("Date")
plt.grid(axis='both',linestyle='-')
plt.legend(loc="upper center")
plt.savefig("./Precipitation in past 12 months.png",bbox_inches='tight')
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
hot_stations = session.query(Measurement.station, Measurement.tobs).order_by(desc(Measurement.tobs)).limit(1)
for hot_station in hot_stations:
    hotest_station = hot_station.station
hotest_station
hotest_station_db = session.query(Measurement.date, Measurement.tobs).\
                filter(Measurement.station==hotest_station).\
                filter(Measurement.date >= one_year_ago).\
                order_by(Measurement.date).all()
hotest_station_db
select_hotest_df = pd.DataFrame(hotest_station_db, columns=['Date', 'Temperature']).set_index('Date')
select_hotest_df['Temperature'].min()
bins = [59.5, 62.5, 65, 67.5, 70, 72.5, 75, 77.5, 80, 82.5, 85, 87.5]
group_names = ["60-62.5","62.5-65", "65-67.5", "67.5-70", "70-72.5", 
               "72.5-75", "75-77.5", "77.5-80", "80-82.5", "82.5-85", "85-87.5"]
select_hotest_df['Temperature Group'] = pd.cut(select_hotest_df['Temperature'],
                                               bins, labels=group_names)
select_hotest_freq_df=select_hotest_df.groupby(['Temperature Group']).count()
select_hotest_freq_df=select_hotest_freq_df.reset_index()
# select_hotest_freq_df=select_hotest_freq_df.set_index('Temperature Group')
select_hotest_freq_df

x_axis = np.arange(len(select_hotest_freq_df))
tick_locations = [value for value in x_axis]
plt.bar(x_axis, select_hotest_freq_df['Temperature'],align='center', label="tobs")
plt.xticks(x_axis, group_names, rotation = 45)
plt.legend(loc="upper right")
plt.ylabel("Frequency")
plt.grid(axis='both',linestyle='-')
plt.title("Temperature in the last 12 months")
plt.savefig("./Temperature in the last 12 months.png",bbox_inches='tight')
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
print(calc_temps('2016-08-23', '2017-08-23'))
print(calc_temps(one_year_ago, last_date))
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

[tmin, tave, tmax]=calc_temps(one_year_ago, last_date)[0]
# print(tmin, tave, tmax)
fig, ax = plt.subplots()
ax.bar("0",tave, width=0.5, alpha=0.5, color = 'coral')
ax.errorbar("0", tave, yerr=(tave-tmin), uplims=True, fmt="o", color="black" )
ax.errorbar("0", tave, yerr=(tmax-tave), lolims=True, fmt="o", color="black" )

ax.set_xlim(-1, 1)
ax.set_title("Trip Avg Temp")
ax.set_ylabel("Temp (F)")
plt.show()
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
start_date="2018-01-01"
end_date="2018-01-07"

# Use the start and end date to create a range of dates
duration=pd.date_range(start=start_date, end=end_date)
# Stip off the year and save a list of %m-%d strings
duration_day=[]
date_range=[]
for date in duration:
    duration_day.append(dt.datetime.strftime(date,"%m-%d"))
    date_range.append(dt.datetime.strftime(date,"%Y-%m-%d"))
duration_day
# Loop through the list of %m-%d strings and calculate the normals for each date
normals_list=[]

for day in duration_day:
    (tmin, tavg, tmax)=daily_normals(day)[0]
    normals_list.append((tmin, tavg, tmax))
print(normals_list)
    
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
normals_list_df=pd.DataFrame(normals_list,columns=['tmin','tavg','tmax'])
normals_list_df['date']=duration
normals_list_df=normals_list_df.set_index('date')
normals_list_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
x_axis = np.arange(len(normals_list_df))
# fig, ax = plt.subplots()
plt.fill_between(x_axis, normals_list_df['tmin'],alpha=0.5, label="tmin", linewidth=1, color="blue")
plt.fill_between(x_axis, normals_list_df['tavg'],alpha=0.3, label="tavg", linewidth=1, color="red")
plt.fill_between(x_axis, normals_list_df['tmax'],alpha=0.3, label="tmax", linewidth=1, color="yellow")
plt.xticks(x_axis, date_range, rotation = 45)
plt.legend(loc="lower left")
plt.xlabel("Date")
plt.show()
```
